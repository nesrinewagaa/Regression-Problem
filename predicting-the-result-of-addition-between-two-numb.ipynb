**Regression problem: Predicting the result of addition between two numbers**

```python

#Import the required python libraries
import numpy as np
from numpy import exp, array, random, dot


#Create the required dataset
x1= np.arange(0, 2, 0.1)
x2= np.arange(0, 2, 0.1)
# Define the input data
X=[[x,y] for x in x1 for y in x2]
X=np.array(X)
# Define the output data
Yd=[[x+y] for x in x1 for y in x2]
Yd = np.array(Yd)
# The user input data to test the efficiency of the implemented network in predicting the desired output
xPredicted = np.array([[0.8,1.8]])


#Create a class called " Neural_Network"
class Neural_Network(object):
  def __init__(self):
    # Initialization of the model weights (W1, W2, and  W3)
    self.W1 = np.random.uniform(size=(2,2))
    self.W2 = np.random.uniform(size=(2,2))
    self.W3 = np.random.uniform(size=(2,1))

    #  Initialization of the model biases (B1, B2,  B3)
    self.B2 =np.random.uniform(size=(1,2))
    self.B1 =np.random.uniform(size=(1,2))
    self.B3 =np.random.uniform(size=(1,1))
    # Define the learning rate

    self.lr=0.0001
  #Define the sigmoid nonlinear activation function

  def sigmoid(self, s):
     return 1/(1+np.exp(-s))

  # Define the derivative of the sigmoid nonlinear activation function

  def sigmoidPrime(self, s):
     return s * (1 - s)

# Define the forward propagation of an ANN model consisting of 3 hidden layers
  def forward(self, X):

     # Result of the first hidden layer
     self.H1 = np.dot(X, self.W1)+ self.B1

     # Apply the sigmoid function to the output of the first hidden layer
     self.Y1 = self.sigmoid(self.H1)

     self.H2 = np.dot(self.Y1, self.W2)+self.B2
     self.Y2 = self.sigmoid(self.H2)
     self.H3 = np.dot(self.Y2, self.W3)+self.B3
     # Result of the third hidden layer
     Y3 = (self.H3)

     return  Y3

#Define the backward propagation of the 3 hidden layers  ANN model
  def backward(self, X, Yd,  Y3):


     # The error between the desired output and the ANN output
     self.Y3_error = Yd - Y3
     self.Y3_delta = self.Y3_error

     self.Y2_error = self.Y3_delta.dot(self.W3.T)
     self.Y2_delta = self.Y2_error*self.sigmoidPrime(self.Y2)

     self.Y1_error = self.Y2_delta.dot(self.W2.T)
     self.Y1_delta = self.Y1_error*self.sigmoidPrime(self.Y1)

     # update W1 and B1
     self.W1 += self.lr*X.T.dot(self.Y1_delta)
     self.B1 += np.sum(self.Y1_delta,axis=0,keepdims=True) * self.lr

     # update W2 and B2
     self.W2  += self.lr * self.Y1.T.dot(self.Y2_delta)
     self.B2= np.sum( self.Y2_delta ,axis=0,keepdims=True) * self.lr

    # update W3 and B3
     self.W3  += self.lr *self.Y2.T.dot(self.Y3_delta)
     self.B3  += np.sum( self.Y3_delta ,axis=0,keepdims=True) * self.lr

# Define the train function

  def train(self, X, Yd):
     Y3 = self.forward(X)
     self.backward(X, Yd, Y3)

# Define the predicted function
  def predict(self):
     print ("Predicted data based on trained weights: ")
     print ("Input : \n" + str(xPredicted))
     print ("Output: \n" + str(self.forward(xPredicted)))

#Train the ANN model and see the predicted result
NN = Neural_Network()
# 100000 is the number of epochs. It defines the number of iterations of how much  running the ANN model
for i in range(100000):
  NN.train(X, Yd)
NN.predict()
```
